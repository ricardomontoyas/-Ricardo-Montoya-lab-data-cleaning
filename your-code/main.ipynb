#### 4. Import posts table:

```python
# We do the same as with users: load the posts data from CSV
posts = pd.read_csv('datasets/posts_table.csv')

# Show first rows to see what columns we have
posts.head()
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
# We rename 'Id' to 'postId' so we know it’s a post ID, not a user ID
# Also rename 'OwnerUserId' to 'userId' so it matches the users table
posts.rename(columns={
    'Id': 'postId',
    'OwnerUserId': 'userId'
}, inplace=True)

posts.head()
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
users = pd.read_csv('datasets/users_table.csv')
users.rename(columns={'Id': 'userId'}, inplace=True)
users.head()
```

```python
# We create a simpler version of the users table
# keeping only the columns we care about

users_clean = users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]

# Same for the posts table, only selecting useful columns
posts_clean = posts[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
# We combine both tables using the userId column
# This adds user information to each post
# how='left' means we keep all the posts, even if the user is missing
merged_df = pd.merge(posts_clean, users_clean, on='userId', how='left')

# Show the first few rows of the merged result
merged_df.head()
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
# This tells us how many missing (NaN) values there are in each column
missing = merged_df.isnull().sum()

# Show only the columns that have at least 1 missing value
missing[missing > 0]
```

#### 10. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
# Make sure that 'userId' and 'postId' are integers
# The type 'Int64' (with capital I) allows for missing values if needed
merged_df['userId'] = merged_df['userId'].astype('Int64')
merged_df['postId'] = merged_df['postId'].astype('Int64')

# List of all the numeric columns we want to make sure are numbers
numeric_cols = ['Score', 'ViewCount', 'CommentCount', 'Reputation', 'Views', 'UpVotes', 'DownVotes']

# Convert each of these columns to numeric type
# errors='coerce' will turn any invalid values into NaN (missing)
merged_df[numeric_cols] = merged_df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Check the final data types of each column
merged_df.dtypes
```

```python
# 'Int64' = integer pero con soporte para valores faltantes.
# 'errors="coerce"' = convierte lo que pueda a número, y lo que no, lo deja como NaN.
```

```python
# ✅ BONUS STEP: Create a dataframe with outliers and export to CSV
# 🧠 ¿Qué es un outlier? Un outlier (valor atípico) es un dato que se sale mucho de lo normal.
# Por ejemplo, si la mayoría de los usuarios tienen reputación entre 1 y 1000, uno con 90000 es un outlier.

# ✅ Paso 1: Elegir una columna para detectar outliers
# Vamos a usar la columna Reputation como ejemplo.

# ✅ Paso 2: Calcular el percentil 99 (top 1%)
# Esto te dará el valor por encima del cual están solo el 1% de los usuarios con más reputación.

# Any value above this threshold will be considered an outlier
reputation_threshold = merged_df['Reputation'].quantile(0.99)

# Show the threshold
print("Outlier threshold for Reputation:", reputation_threshold)
```

```python
#✅ Paso 3: Crear un nuevo DataFrame con solo los outliers
# Create a new DataFrame with only the outliers
outliers_df = merged_df[merged_df['Reputation'] > reputation_threshold]

# Show a preview
outliers_df.head()
```

```python
# ✅ Paso 4: Export the outliers DataFrame to a CSV file in the 'your-code' folder
# Make sure you already have the outliers_df created
# (If not, run the outlier detection step first)

# Save the file
outliers_df.to_csv('weather.ipynb', index=False)

# Confirm it was saved (optional)
print("✅ Outliers exported successfully to weather.ipynb")
```
